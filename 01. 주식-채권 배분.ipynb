### Yfinance로 ETF 가격 데이터 다운로드

```python
tickers = ['SPY', 'TLT']
etf = yf.Tickers(tickers)

data = etf.history(period='max')
data
```

```python
data.drop(['Open', 'Low', 'High', 'Volume', 'Dividends', 'Stock Splits', 'Capital Gains'], inplace=True, axis=1)
data
```

```python
data = data.droplevel(0, axis=1)
data
```

```python
data.plot()
```

### 입력변수 : 수익률, 변동성, 상관계수

#### 1. 수익률

```python
# 수익률 계산
rets = data.pct_change().fillna(0)

# 수익률 데이터 확인
rets.head()

# 수익률 시각화
rets.plot()
plt.title("Historical returns")
plt.xlabel("Date")
plt.ylabel("Returns")
```

#### 2. 변동성

```python
import numpy as np

# 전체 기간 변동성 계산
total_vols = rets.std()*np.sqrt(252)
print(total_vols)

# 1년 단위 롤링 변동성 계산 및 연율화
vols = rets.rolling(252).std()*np.sqrt(252)

# 변동성 시각화
vols.plot()
plt.title("Historical vols")
plt.xlabel("Date")
plt.ylabel("Vols")
```

#### 3. 상관계수

```python
# SPY / TLT 수익률
spy_rets = rets['SPY']
tlt_rets = rets['TLT']

# 전체 기간 상관계수 계산
total_corr = spy_rets.corr(tlt_rets)
print(f"전체 기간 SPY-TLT 상관계수 : {total_corr}")

# 1년 단위 롤링 상관계수 계산
corr = spy_rets.rolling(252).corr(tlt_rets)

# 상관계수 시각화
corr.plot()
plt.title("Historical Correlation")
plt.xlabel("Date")
plt.ylabel("Correlation")
```

```python
# [ETF Universe]
# Ticker / Sector
# XLB / 소재
# XLE / 에너지
# XLF / 금융
# XLI / 산업재
# XLK / IT
# XLP / 필수소비재
# XLU / 유틸리티
# XLV / 헬스케어
# XLY / 임의소비재
```

```python
# ETF Universe Ticker 설정
tickers = ['XLB', 'XLE', 'XLF', 'XLI', 'XLK', 'XLP', 'XLU', 'XLV', 'XLY']

# Ticker 객체 선언
yf_obj = yf.Tickers(tickers)

# 과거 데이터 다운로드
df = yf_obj.history(period='max')

# 종가를 제외한 나머지 컬럼 삭제
df.drop(['Open', 'Low', 'High', 'Volume', 'Dividends', 'Stock Splits', 'Capital Gains'], inplace=True, axis=1)

# NaN 값을 이전 값으로 채워넣기
df.ffill(inplace=True)

# 첫번째 칼럼 레벨 삭제
df = df.droplevel(0, axis=1)

# 주간 단위로 데이터 재설정
df = df.resample('W').last()
```

```python
df
```

```python
df.plot()
```

```python
# ETF 수익률 계산
rets = df.pct_change().fillna(0)

# 수익률 시각화
rets.plot(figsize=(12, 8))
```

```python
# 1년 단위 롤링 변동성 계산 및 연율화
vols = rets.rolling(52).std()*np.sqrt(52)

# 변동성 시각화
vols.plot(figsize=(12, 8))
```

```python
# 전체 기간 ETF 상관계수 계산
corr = rets.corr()
corr
```

```python
# seaborn 패키지 임포트
import seaborn as sns

# 히트맵 시각화
plt.figure(figsize=(15, 8))
sns.heatmap(corr, annot=True)
```

#### 공분산 행렬 특징 이해하기

```python
# 공분산 계산 및 연율화
cov = rets.cov()*52
cov
```

```python
# var() 메서드를 사용한 ETF 분산 계산
vols_1 = rets.var()*52
vols_1
```

```python
# 공분산 행렬 및 diag() 메서드를 사용한 ETF 분산 계산
vols_2 = np.diag(cov)
vols_2
```

### 주식-채권 배분 모델

```python
# 패키지 임포트
import numpy as np
import yfinance as yf
import seaborn as sns
sns.set()
```

#### 가격 데이터 다운로드 및 전처리

```python
# 티커
tickers = ['SPY', 'TLT']

# 티커 객체
etf = yf.Tickers(tickers)

# 데이터 다운로드
data = etf.history(start='2002-08-01', actions=False)
```

```python
# 불필요한 칼럼 제거
data.drop(['Open', 'Low', 'High', 'Volume'], inplace=True, axis=1)
```

```python
# 0 레벨 컬럼 제목 제거
data = data.droplevel(0, axis=1)
```

```python
# 가격 그래프
data.plot()
```

#### 일별 수익률 데이터

```python
# 일별 수익률 데이터
rets = data.pct_change().fillna(0)

# 수익률 데이터 시각화
rets.plot()
```

```python
rets
```

#### 개별 자산 평균 수익률과 변동성, 샤프 비율

```python
# 주식 기대 수익률과 변동성
stock_mu = rets['SPY'].mean()*252
stock_sigma = rets['SPY'].std()*np.sqrt(252)

# 주식 샤프 비율
stock_sr = stock_mu/stock_sigma

# 채권 기대 수익률과 변동성
bond_mu = rets['TLT'].mean()*252
bond_sigma = rets['TLT'].std()*np.sqrt(252)

# 채권 샤프 비율
bond_sr = bond_mu/bond_sigma

# 결과 출력
print(f"SPY 수익률 : {stock_mu}, TLT 수익률 : {bond_mu}")
print(f"SPY 변동성 : {stock_sigma}, TLT 변동성 : {bond_sigma}")
print(f"SPY 샤프 비율 : {stock_sr}, TLT 사프 비율 : {bond_sr}")
```

#### 주식-채권 60:40 포트폴리오 구현

```python
# 주식 가중치 60%, 채권 가중치 40%
weights = 0.7
```

```python
# 포트폴리오 수익률
port_ret = weights*rets['SPY'] + (1-weights)*rets['TLT']
```

```python
# NaN 값 제거
port_ret.dropna(inplace=True)
```

```python
# 결과 출력
port_ret
```

```python
# 그래프 시각화
(1+port_ret).cumprod().plot()
(1+rets['SPY']).cumprod().plot()
(1+rets['TLT']).cumprod().plot()
```

#### 포트폴리오 수익률과 변동성 분석

```python
# 포트폴리오 기대 수익률
port_mu = port_ret.mean()*252

# 포트폴리오 변동성
port_sigma = port_ret.std()*np.sqrt(252)

# 결과 출력
print(f"Returns = {port_mu}")
print(f"Vol = {port_sigma}")
```

#### 주식-채권 상관계수 분석

```python
# 1년 롤링 상관계수
historical_corr = rets['SPY'].rolling(252).corr(rets['TLT'])
```

```python
# 롤링 상관계수 시각화
historical_corr.plot()
```
